```python
hard_coded_url = ['https://www.tesco.com/groceries/en-GB/shop/fresh-food/all?page={}&count=24',
                  'https://www.tesco.com/groceries/en-GB/shop/frozen-food/all?page={}&count=24',
                  'https://www.tesco.com/groceries/en-GB/shop/drinks/all?page={}&count=24',
                  'https://www.tesco.com/groceries/en-GB/shop/baby/all?page={}&count=24',
                  'https://www.tesco.com/groceries/en-GB/shop/household/all?page={}&count=24',
                  'https://www.tesco.com/groceries/en-GB/shop/food-cupboard/all?page={}&count=24']

for i in hard_coded_url:
    for j in range(300):
        page_num = j + 1
        url = i.format(page_num)
        r = requests.get(url , headers = headers)
        if r.status_code == 200:
            url_list.append(url)
        elif r.status_code != 200:
            break
```

```python
for j in url_list:
    r = requests.get(j , headers = headers)
    soup = BeautifulSoup(r.content , 'html.parser')
    productscat = soup.find('h1',{'class':'heading query'})
    rows = soup.find_all('div',{'class':'product-details--wrapper'})
    for row in rows:
        products_category.append(productscat.text)
        product = row.find('span',{'class':'styled__Text-sc-1xbujuz-1 ldbwMG beans-link__text'})
        products_list.append(product.text)
        if  row.find('p',{'class':'styled__StyledHeading-sc-119w3hf-2 jWPEtj styled__Text-sc-8qlq5b-1 lnaeiZ beans-price__text'}) is None:
            price = 'Out of Stock'
        else:
            price = (row.find('p',{'class':'styled__StyledHeading-sc-119w3hf-2 jWPEtj styled__Text-sc-8qlq5b-1 lnaeiZ beans-price__text'}))
            price = price.text
        price_list.append(price)
```
